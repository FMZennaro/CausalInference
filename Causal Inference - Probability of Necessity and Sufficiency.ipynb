In this notebook we empirically evaluate probability of necessity and sufficiency.

### Models
We consider three different *structural causal models*, all having the same underlying *DAG*, but differing in the *structural equations*.

**DAG.** We consider the following simple graph, made up by three deterministic *endogenous variables*: a treatment variable ($X$), a confounder variable ($Z$) and an outcome variable ($Y$). We also assume that a random *exogenous variable* is connected to each endogenous node ($U_X, U_Z, U_Y$).

**Structural equations.** All the models share the same equations for the *exogenous nodes* ($U_Z, U_X, U_Y$) and for two of the three *endogenous nodes* ($Z, X$); the main difference is in the definition of the structural equation for the last *endogenous node* ($Y$)

$$
\begin{array}{ccc}
 & U_{Z}=\mathtt{Unif}\left(0,1\right)\\
 & U_{X}=\mathtt{Bern}\left(0.5\right)\\
 & U_{Y}=\mathtt{Pois}\left(2\right)\\
\\
 & Z=U_{Z}\\
 & X=1\left[U_{X}+Z\geq1.2\right]\\
Y^{M1}=1\left[Xe^{Z}U_{Y}\geq1\right] & \,\,\,Y^{M2}=1\left[2X-Z+U_{Y}\geq1\right]\,\,\, & Y^{M3}=1\left[\frac{X}{2}-Z+U_{Y}\geq1\right]
\end{array}
$$

Notice that we use the notation $A = 1\left[c\right]$ as a shorthand for the conditional over $c$: $A=1$ if $c$ is true, $A=0$ otherwise.

## Setup

Importing libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import seaborn as sns

import warnings
```

Setting parameters for running inference algorithms.

```python
n_samples=5000
```

## Sampling the models
We start by drawing random samples from the models to observe the distributions $P(X), P(Z)$ and $P(Y)$.

We define each model in a function *model_()* that instantiates a PyMC3 model and samples from it. We then plot the distribution for the variables of interest.

```python
def model1():
    with pm.Model():
        Uz = pm.Uniform('Uz',0,1)
        Ux = pm.Bernoulli('Ux',p=.5)
        Uy = pm.Poisson('Uy',mu=2)
         
        Z = pm.Deterministic('Z', Uz)
        X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
        Y = pm.Deterministic('Y', pm.math.ge(X*pm.math.exp(Z)*Uy,1))
                
        trace = pm.sample(n_samples)
        return trace['X'],trace['Y'], trace['Z']
     
X,Y,Z = model1()
fig, axes = plt.subplots(1,3, figsize=(10,4))
fig.suptitle('Model 1')
axes[0].set_title('P(X)')
sns.distplot(np.array(X,dtype=np.float32), kde=False, color='green', ax=axes[0]) 
axes[1].set_title('P(Z)')
sns.distplot(Z, kde=False, color='blue', ax=axes[1])
axes[2].set_title('P(Y)')
sns.distplot(np.array(Y,dtype=np.float32), kde=False, color='red', ax=axes[2])
```

```python
def model2():
    with pm.Model():
        Uz = pm.Uniform('Uz',0,1)
        Ux = pm.Bernoulli('Ux',p=.5)
        Uy = pm.Poisson('Uy',mu=2)
         
        Z = pm.Deterministic('Z', Uz)
        X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
        Y = pm.Deterministic('Y', pm.math.ge(2*X - Z + Uy,1))
                
        trace = pm.sample(n_samples)
        return trace['X'],trace['Y'], trace['Z']
     
X,Y,Z = model2()
fig, axes = plt.subplots(1,3, figsize=(10,4))
fig.suptitle('Model 2')
axes[0].set_title('P(X)')
sns.distplot(np.array(X,dtype=np.float32), kde=False, color='green', ax=axes[0]) 
axes[1].set_title('P(Z)')
sns.distplot(Z, kde=False, color='blue', ax=axes[1])
axes[2].set_title('P(Y)')
sns.distplot(np.array(Y,dtype=np.float32), kde=False, color='red', ax=axes[2])
```

```python
def model3():
    with pm.Model():
        Uz = pm.Uniform('Uz',0,1)
        Ux = pm.Bernoulli('Ux',p=.5)
        Uy = pm.Poisson('Uy',mu=2)
         
        Z = pm.Deterministic('Z', Uz)
        X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
        Y = pm.Deterministic('Y', pm.math.ge(X/2. - Z + Uy,1))
                
        trace = pm.sample(n_samples)
        return trace['X'],trace['Y'], trace['Z']
     
X,Y,Z = model3()
fig, axes = plt.subplots(1,3, figsize=(10,4))
fig.suptitle('Model 2')
axes[0].set_title('P(X)')
sns.distplot(np.array(X,dtype=np.float32), kde=False, color='green', ax=axes[0]) 
axes[1].set_title('P(Z)')
sns.distplot(Z, kde=False, color='blue', ax=axes[1])
axes[2].set_title('P(Y)')
sns.distplot(np.array(Y,dtype=np.float32), kde=False, color='red', ax=axes[2])
```

In all cases we note that the discrete pmf of $P(X)$ is slightly skewed towards 0 and that the pdf of $P(Z)$ is roughly uniform. The pmf of $P(Y)$ varies according to the model.

## Probability of necessity
Proability of necessity (PN) is defined as:
$$
P\left( Y_0=0 \vert X=1, Y=1 \right),
$$
that is, the *counterfactual* probability that $Y$ would be $0$ if we were to set $X=0$, given that in the *factual* world we actually observed $X=1$ and $Y=1$. In the medical setting, it is the probability that the outcome would have been negative if we had not applied the treatment, knowing that, in fact, the treatment was applied and the outcome was positive. Thus, the probability of necessity measures the **strength of necessity of X=1 in causing Y=1, conditioned on the fact that we observed X=1, Y=1**; this quantity is evaluated in a *counterfactual world*, where we compute the probability that $Y=0$ if we had set $X=0$.

Again we define each model in a function *model_PN()* that instantiates a PyMC3 model; we then generate individual samples from the model inside a *while* loop, until we obtain a sample such that $(X=1,Y=1)$; at this point we use the sample to evaluate the counterfactual $Y_0$ under the intervention $do(X=0)$. We repeat these procedure for each sample and then, finally, we plot the distribution of the probability of necessity $P\left( Y_0=0 \vert X=1, Y=1 \right)$.

Notice that this procedure is highly inefficient, since we set up the whole pymc3 inference system (which, to the best of my knowledge, does not implement simple forward sampling) just to get a single sample; it would be more efficent to sample a batch of data samples and work on them. However, it seems to me that the single-sample procedure is conceptually closer to the individual-level evaluation performed by counterfactual reasoning. Thus, even if inefficient, the single-sample procedure is preferred here because closer to the intuition. Notice that before running inference warnings are disabled to avoid printing pymc3 warning about too small a number of draws.

### Model 1

```python
def model1_PN():
    Yfactual = False
    Xfactual = False
     
    while not(Yfactual==True and Xfactual==True):    
        with pm.Model():
            Uz = pm.Uniform('Uz',0,1)
            Ux = pm.Bernoulli('Ux',p=.5)
            Uy = pm.Poisson('Uy',mu=2)
             
            Z = pm.Deterministic('Z', Uz)
            X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
            Y = pm.Deterministic('Y', pm.math.ge(X*pm.math.exp(Z)*Uy,1))
            
            warnings.filterwarnings("ignore")
            trace = pm.sample(1, chains=1, progressbar = False)
            warnings.filterwarnings("default")
            
            #print(trace['X'][0])
            #print(trace['Y'][0])
            Yfactual=trace['X'][0]  
            Xfactual=trace['Y'][0]
     
    Uz_counterfactual = trace['Z'][0]
    Ux_counterfactual = trace['Ux'][0]
    Uy_counterfactual = trace['Uy'][0]
     
    Z_counterfactual = trace['Z'][0]
    X_counterfactual = 0.0
    Y_counterfactual = X_counterfactual*np.exp(Z_counterfactual)*Uy_counterfactual >= 1
            
    return Y_counterfactual
 
PN_Y = []
for i in range(10):
    #print('Collecting {0}th datasample'.format(i+1))
    PN_Y.append(model1_PN())
plt.figure()
plt.title('Proability of necessity of X in model 1')
sns.distplot(np.array(PN_Y,dtype=np.float32), kde=False, bins=[-.2,.2,.8,1.2], color='red')
```

The pmf for the probability of necessity is identically 0. This means that if we had (countefactually) intervened on $do(X=0)$, when $X=1,Y=1$, then the outcome $Y$ would have certainly changed to 0. In other words, when $X=1,Y=1$, then $X$ being 1 is necessary for $Y$ being 1. In the medical setting, if we consider a patient who had the treatment and had a positive outcome, we can say that having the treatment was necessary for his having a positive outcome.
The result we obtained is consistent with the structural equation for $Y^{M1}=1\left[Xe^{Z}U_{Y}\geq1\right]$; indeed this condition can not evaluate to 1 if $X=0$.

Notice that this does **NOT** mean in general that $X=1$ causes $Y=1$ nor that $Y=1$ always implies $X=1$; there may be factual cases where $X=0,Y=1$ and $X=1,Y=0$. All our conclusions on the necessity of $X=1$ are relative to the factual world in which $X=1,Y=1$.

Thus, with reference to the medical setting, given $P\left( Y_0=0 \vert X=1, Y=1 \right) = a$, in summary we can say:
<li>For a patient who received a treatment and had a positive output, treatment was $a\%$ necessary for recovery</li>
but we can **NOT** say:
<li>Treatment was overall $a\%$ necessary for recovery; there may be cases where a patient did not receive the treatment ($X=0$) but recovered ($Y=1$) thanks to some confounding genetic disposition ($Z$) or cases in which a patient received the treatment ($X=1$) but did not recover ($Y=0$) due to some confounding health condition ($Z$) </li> 

Also notice that the probality of necessity $P\left( Y_0=0 \vert X=1, Y=1 \right)$ is **NOT** necessarily the same as the conditional distribution $P\left( Y=0 \vert X=0 \right)$. In same cases they may be the same, as with the current *model1* shown below. (Notice that in this code we artificially substitute a *Deterministic* node with a *Normal* ($mu, 0.001$) node to allow pymc3 to run its inference).

```python
def model1_condX0():
    Yfactual = False
    Xfactual = False
     
    with pm.Model():
        Uz = pm.Uniform('Uz',0,1)
        Ux = pm.Bernoulli('Ux',p=.5)
        Uy = pm.Poisson('Uy',mu=2)

        Z = pm.Deterministic('Z', Uz)
        X0 = pm.Normal('X0', mu=Ux+Z, sd=0.001, observed=0)
        X = pm.Deterministic('X', pm.math.ge(X0, 1.2))
        Y = pm.Deterministic('Y', pm.math.ge(X*pm.math.exp(Z)*Uy,1))

        trace = pm.sample(chains=1)

    return trace['Y']
 
condX0_Y = model1_condX0()
plt.figure()
plt.title('Proability of Y given X=0')
sns.distplot(np.array(condX0_Y,dtype=np.float32), kde=False, bins=[-.2,.2,.8,1.2], color='red')
```

However, in general this is not the case that the *probability of necessity* and *conditional distribution* are the same. In the following case, for instance, the two quantities would differ:
$$
\begin{array}{ccc}
X & Z & Y=f(X,Z)\\
1 & 1 & 1\\
1 & 0 & 0\\
0 & 1 & 1\\
0 & 0 & 0
\end{array}
$$

### Model 2

```python
def model2_PN():
    Yfactual = False
    Xfactual = False
     
    while not(Yfactual==True and Xfactual==True):    
        with pm.Model():
            Uz = pm.Uniform('Uz',0,1)
            Ux = pm.Bernoulli('Ux',p=.5)
            Uy = pm.Poisson('Uy',mu=2)
         
            Z = pm.Deterministic('Z', Uz)
            X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
            Y = pm.Deterministic('Y', pm.math.ge(2*X - Z + Uy,1))
            
            warnings.filterwarnings("ignore")
            trace = pm.sample(1, chains=1, progressbar = False)
            warnings.filterwarnings("default")
            
            #print(trace['X'][0])
            #print(trace['Y'][0])
            Yfactual=trace['X'][0]  
            Xfactual=trace['Y'][0]
     
    Uz_counterfactual = trace['Z'][0]
    Ux_counterfactual = trace['Ux'][0]
    Uy_counterfactual = trace['Uy'][0]
     
    Z_counterfactual = trace['Z'][0]
    X_counterfactual = 0.0
    Y_counterfactual = X_counterfactual*2 - Z_counterfactual + Uy_counterfactual >= 1
            
    return Y_counterfactual
 
PN_Y = []
for i in range(25):
    #print('Collecting {0}th datasample'.format(i+1))
    PN_Y.append(model2_PN())
plt.figure()
plt.title('Proability of necessity of X in model 2')
sns.distplot(np.array(PN_Y,dtype=np.float32), kde=False, bins=[-.2,.2,.8,1.2], color='red')
```

The pmf for the probability distributes over 0 and 1. The pmf is built using only 25 samples, so it is not very reliable, but still useful in order to make some observations. If we had (countefactually) intervened on $do(X=0)$, when $X=1,Y=1$, then the outcome $Y$ would could have stayed the same at the value $1$ or changed to $0$. In other words, when $X=1,Y=1$, then $X$ being $1$ is not strictly necessary for $Y$ being 1. In the medical setting, if we consider a patient who had the treatment and had a positive outcome, we can say that having the treatment was not strictly necessary for his having a positive outcome. Actually we could quantify this degree of necessity via a better estimation (more samples) of this probability of necessity. 
The result we obtained is consistent with the structural equation for $Y^{M2}=1\left[2X-Z+U_{Y}\geq1\right]$; indeed the condition can sometimes evaluate to $1$ if thanks to $X=1$, but in some circumstances it can ealuate to $1$ if thanks to $U_{Y}=1$.

(All the observations above on the interpretation of the probability of necessity holds here as well).

### Model 3

```python
def model3_PN():
    Yfactual = False
    Xfactual = False
     
    while not(Yfactual==True and Xfactual==True):    
        with pm.Model():
            Uz = pm.Uniform('Uz',0,1)
            Ux = pm.Bernoulli('Ux',p=.5)
            Uy = pm.Poisson('Uy',mu=2)
         
            Z = pm.Deterministic('Z', Uz)
            X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
            Y = pm.Deterministic('Y', pm.math.ge(X/2. - Z + Uy,1))
            
            warnings.filterwarnings("ignore")
            trace = pm.sample(1, chains=1, progressbar = False)
            warnings.filterwarnings("default")
            
            #print(trace['X'][0])
            #print(trace['Y'][0])
            Yfactual=trace['X'][0]  
            Xfactual=trace['Y'][0]
     
    Uz_counterfactual = trace['Z'][0]
    Ux_counterfactual = trace['Ux'][0]
    Uy_counterfactual = trace['Uy'][0]
     
    Z_counterfactual = trace['Z'][0]
    X_counterfactual = 0.0
    Y_counterfactual = X_counterfactual/2. - Z_counterfactual + Uy_counterfactual >= 1
            
    return Y_counterfactual
 
PN_Y = []
for i in range(25):
    #print('Collecting {0}th datasample'.format(i+1))
    PN_Y.append(model3_PN())
plt.figure()
plt.title('Proability of necessity of X in model 3')
sns.distplot(np.array(PN_Y,dtype=np.float32), kde=False, bins=[-.2,.2,.8,1.2], color='red')
```

The pmf for the probability of necessity seems to be strongly concentrad over 1 (with the caveat that, again, we used only 25 samples to evaluate this pmf). This result means that if we had (countefactually) intervened on $do(X=0)$, when $X=1,Y=1$, then the outcome $Y$ would have very likely stayed the same at the value $1$. In other words, when $X=1,Y=1$, then $X$ being $1$ is almost completely unnecessary for $Y$ being 1. In the medical setting, if we consider a patient who had the treatment and had a positive outcome, we can say that having the treatment was almost certainly superfluous for his having a positive outcome. An exact estimation of the probability of necessity (collecting more samples) would allow for a rigorous quantification of the statement "almost certainly superfluous".
The result we obtained is consistent with the structural equation for $Y^{M3}=1\left[\frac{X}{2}-Z+U_{Y}\geq1\right]$; indeed, under this model, we can expect the contribution of $X=1$ to $Y=1$ to be very limited.

(All the observations above on the interpretation of the probability of necessity holds here as well).

## Probability of sufficiency
Proability of sufficiency (PS) is defined as:
$$
P\left( Y_1=1 \vert X=0, Y=0 \right),
$$
that is, the *counterfactual* probability that $Y$ would be $1$ if we were to set $X=1$, given that in the *factual* world we actually observed $X=0$ and $Y=0$. In the medical setting, it is the probability that the outcome would have been positive if we had applied the treatment, knowing that, in fact, the treatment was not applied and the outcome was negative. Thus, the probability of sufficiency measures the **strength of sufficiency of X=1 in causing Y=1, conditioned on the fact that we observed X=0, Y=0**; this quantity is evaluated in a *counterfactual world*, where we commpute the probability that $Y=1$ if we had set $X=1$.

Similarly to what we did for the probability of necessity, we define each model in a function *model_PS()* that instantiates a PyMC3 model; we then generate individual samples from the model inside a *while* loop, until we obtain a sample such that $(X=0,Y=0)$; at this point we use the sample to evaluate the counterfactual $Y_1$ under the intervention $do(X=1)$. We repeat these procedure for each sample and then, finally, we plot the distribution of the probability of necessity $P\left( Y_1=1 \vert X=0, Y=0 \right)$.

### Model 1

```python
def model1_PS():
    Yfactual = True
    Xfactual = True
     
    while not(Yfactual==False and Xfactual==False):    
        with pm.Model():
            Uz = pm.Uniform('Uz',0,1)
            Ux = pm.Bernoulli('Ux',p=.5)
            Uy = pm.Poisson('Uy',mu=2)
             
            Z = pm.Deterministic('Z', Uz)
            X = pm.Deterministic('X', pm.math.ge(Ux+Z, 1.2))
            Y = pm.Deterministic('Y', pm.math.ge(X*pm.math.exp(Z)*Uy,1))
            
            warnings.filterwarnings("ignore")
            trace = pm.sample(1, chains=1, progressbar = False)
            warnings.filterwarnings("default")
            
            #print(trace['X'][0])
            #print(trace['Y'][0])
            Yfactual=trace['X'][0]  
            Xfactual=trace['Y'][0]
     
    Uz_counterfactual = trace['Z'][0]
    Ux_counterfactual = trace['Ux'][0]
    Uy_counterfactual = trace['Uy'][0]
     
    Z_counterfactual = trace['Z'][0]
    X_counterfactual = 1.0
    Y_counterfactual = X_counterfactual*np.exp(Z_counterfactual)*Uy_counterfactual >= 1
            
    return Y_counterfactual
 
PS_Y = []
for i in range(25):
    #print('Collecting {0}th datasample'.format(i+1))
    PS_Y.append(model1_PS())
plt.figure()
plt.title('Proability of sufficiency of X in model 1')
sns.distplot(np.array(PS_Y,dtype=np.float32), kde=False, bins=[-.2,.2,.8,1.2], color='red')
```

The pmf for the probability of sufficiency distributes mainly over 0. This means that if we had (countefactually) intervened on $do(X=1)$, when $X=0,Y=0$, then the outcome $Y$ would have with high probability changed to 1. In other words, when $X=0,Y=0$, then setting $X=1$ is likely sufficient for $Y$ being $1$. In the medical setting, if we consider a patient who had not received the treatment and had a negative outcome, we can say that having the treatment would have very likely caused a positive outcome. Again, a proper estimation of the probability of sufficiency allows us to quantify statements like "would have very likely".
The result we obtained is consistent with the structural equation for $Y^{M1}=1\left[Xe^{Z}U_{Y}\geq1\right]$; indeed, $Y=0$ is likely determined by $X=0$ alone, and only in limited cases we may expect to have $X=0$ and $U_{Y}$; thus switching $X$ to $1$ would likely change the value of $Y$ to $1$.

### Model 2

The pmf for the probability of sufficiency is identically 1. This means that if we had (countefactually) intervened on $do(X=1)$, when $X=0,Y=0$, then the outcome $Y$ would have certainly changed to the value $1$. In other words, when $X=0,Y=0$, then setting $X$ to $1$ is sufficient for $Y$ being 1. The result we obtained is consistent with the structural equation for $Y^{M2}=1\left[2X-Z+U_{Y}\geq1\right]$; indeed the condition evaluates to $1$ whenever $X=1$.
